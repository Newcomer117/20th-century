Import libraries

```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

```python
# Import txt file

with open('Key_events_of_the_20th_century.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
# Sentence tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent) 
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word) 
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
dist_words.most_common(10)
```

```python
top10 = dist_words.most_common(10)
words = [w for w, c in top10]
counts = [c for w, c in top10]

plt.figure(figsize=(8, 3))
plt.bar(words, counts)
plt.show()
```

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')  # run once

stop_words = set(stopwords.words('english'))
```

```python
# Removing stopwords in words
filtered_words = []  # creates an empty list

for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
# Create a new FreqDist for filtered–words

dist_words_filter = FreqDist(filtered_words)
print (dist_words_filter)
```

```python
# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print (tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2) 
```

```python
# Get 10 most common words from the filtered distribution
top10 = dist_words_filter_2.most_common(10)
words = [w for w, c in top10]
counts = [c for w, c in top10]

# Bar chart
plt.figure(figsize=(8, 3))
plt.bar(words, counts)
plt.show()
```

Now that the punctioation and filler words have been removed, it is actually possible to recognize what the text is about.

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The']

filtered = []
for word in tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top20=df_t.nlargest(20, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top20).set_title("Key_events_of_the_20th_century - top 20 word types used")
```

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences')
```

```python
top15 = y.nlargest(15, 'Occurrences')

plt.figure(figsize=(12, 5))
sns.barplot(
    x="Word",
    y="Occurrences",
    data=top15,
    palette="Blues",
    hue="Word",
    legend=False
).set_title("Key_events_of_the_20th_century - most frequently used nouns")
```

```python
# Filter for verbs
df_verbs = df_text[df_text['Word type'].isin(["VB", "VBD", "VBG", "VBN", "VBP", "VBZ"])]
df_verbs.columns = ["Word", "Occurrences"]

# Count occurrences per verb
x_verbs = df_verbs.groupby('Word').count().reset_index()
y_verbs = x_verbs.sort_values(by='Occurrences', ascending=False)

# Top 15 verbs
top15_verbs = y_verbs.nlargest(15, 'Occurrences')

# Bar chart
plt.figure(figsize=(12, 5))
sns.barplot(
    x="Word",
    y="Occurrences",
    data=top15_verbs,
    palette="Greens",
    hue="Word",
    legend=False
).set_title("Key_events_of_the_20th_century - most frequently used verbs")
```

```python
# Filter for adjectives
df_adj = df_text[df_text['Word type'].isin(["JJ", "JJR", "JJS"])]
df_adj.columns = ["Word", "Occurrences"]

# Count occurrences per adjective
x_adj = df_adj.groupby('Word').count().reset_index()
y_adj = x_adj.sort_values(by='Occurrences', ascending=False)

# Top 15 adjectives
top15_adj = y_adj.nlargest(15, 'Occurrences')

# Bar chart
plt.figure(figsize=(12, 5))
sns.barplot(
    x="Word",
    y="Occurrences",
    data=top15_adj,
    palette="Oranges",
    hue="Word",
    legend=False
).set_title("Key_events_of_the_20th_century - most frequently used adjectives")
```

Having the most common words for nouns, adjectives and verbs gives a clearer view on what the text is about. Someone not knowing the text, could now guess the topic.
