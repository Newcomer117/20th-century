```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Line 1: create Service object
service = Service(ChromeDriverManager().install())

# Line 2: create Chrome driver using the service
driver = webdriver.Chrome(service=service)
```

```python
# Get the pageâ€™s contents

page_url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century"
driver.get(page_url)
```

```python
from bs4 import BeautifulSoup
import requests 
```

```python
# Get URL

page_url =  requests.get("https://en.wikipedia.org/wiki/Key_events_of_the_20th_century")
```

```python
# Create soup and get title

soup = BeautifulSoup(page_url.text, 'html.parser')
print(soup.title) 
```

```python
text = soup.get_text()

text = text.encode ('utf-8')

with open('Key_events_of_the_20th_century.txt', 'wb') as f:
       f.write(text)
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}

page_url = requests.get(
    "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century",
    headers=headers
)
```

```python
soup = BeautifulSoup(page_url.text, "html.parser")
text = soup.get_text()
```

```python
text = text.encode("utf-8")

with open("Key_events_of_the_20th_century.txt", "wb") as f:
    f.write(text)
```
